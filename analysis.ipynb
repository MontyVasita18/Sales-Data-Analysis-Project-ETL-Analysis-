# Writing Querys

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Show Top 5 Countries 

```python
query = """
SELECT Country, COUNT(*) AS Total_Customers 
FROM Customers 
GROUP BY Country 
ORDER BY Total_Customers DESC 
LIMIT 5;
"""

top_countries = pd.read_sql_query(query, engine)
top_countries
```

```python
plt.pie(top_countries['Total_Customers'],
        labels=top_countries['Country'],
        autopct='%1.1f%%',

)
plt.show()
```

2. Show the top 5 most expensive products.

```python
query = """
SELECT ProductName, UnitPrice 
FROM Products
ORDER BY UnitPrice DESC
LIMIT 5;
"""

top_expensive_products = pd.read_sql_query(query, engine)
top_expensive_products
```

```python
sns.barplot(data=top_expensive_products,x='UnitPrice',y='ProductName')
plt.show()
```

3. Display the total number of orders per shipping country.

```python
query = """
SELECT ShipCountry, COUNT(OrderID) AS TotalOrders
FROM Orders
GROUP BY ShipCountry
ORDER BY TotalOrders DESC;
"""

order_by_Country=pd.read_sql(query,engine)
order_by_Country
```

```python
plt.pie(order_by_Country['TotalOrders'],
        labels=order_by_Country['ShipCountry'],
        autopct='%1.1f%%',
        startangle=140,
)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.tight_layout()
plt.show()
```

4. Most ordered Product

```python
query = """
 select
    p.ProductName,
    SUM(o.Quantity) AS TotalOrdered
FROM Orders o
JOIN Products p ON o.ProductID = p.ProductID
GROUP BY p.ProductID
ORDER BY TotalOrdered DESC
LIMIT 5;
"""

most_ordered_product = pd.read_sql(query, engine)
most_ordered_product

```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=most_ordered_product,x='ProductName',y='TotalOrdered')
plt.xlabel('Product Name')
plt.ylabel('Total_Ordered')
plt.show()
```

5. Employees with Most Orders

```python
 # Employees with Most Orders

query="""
select e.EmployeeID, e.FirstName, e.LastName, COUNT(o.OrderID) AS OrdersProcessed
FROM Employees e
JOIN Orders o ON e.EmployeeID = o.EmployeeID
GROUP BY e.EmployeeID
ORDER BY OrdersProcessed DESC;
"""
employee_orders = pd.read_sql(query, engine)
employee_orders
```

5. Get order details with product name, quantity, discount, and final price.

```python
query = """
Select 
    o.OrderID,
    p.ProductName,
    o.Quantity,
    o.UnitPrice,
    o.Discount,
    (o.UnitPrice * o.Quantity * (1 - o.Discount / 100)) AS FinalPrice
FROM Orders o
JOIN Products p ON o.ProductID = p.ProductID;
"""

order_details = pd.read_sql(query, engine)
order_details
```

6. Customers With No Orders

```python
query = """
Select c.CustomerID, c.CompanyName
FROM Customers c
LEFT JOIN Orders o ON c.CustomerID = o.CustomerID
WHERE o.OrderID IS NULL;
"""

customers_without_orders = pd.read_sql(query, engine)
customers_without_orders.head()
```

7. Show all products that are discontinued

```python

query="""
select ProductID,ProductName
from products
where Discontinued= 1;
"""
Discontinue_products=pd.read_sql(query,engine)
Discontinue_products
```

8. Find most frequent customers who have placed more than 5 orders.

```python
query = """
SELECT CompanyName
FROM Customers
WHERE CustomerID IN (
    SELECT CustomerID
    FROM Orders
    GROUP BY CustomerID
    HAVING COUNT(OrderID) > 5
);
"""

frequent_customers = pd.read_sql(query, engine)
frequent_customers.head()
```

9.Find the most ordered product (based on quantity)

```python
query = """
SELECT 
    p.ProductName,
    SUM(o.Quantity) AS TotalOrdered
FROM Orders o
JOIN Products p ON o.ProductID = p.ProductID
GROUP BY p.ProductID
ORDER BY TotalOrdered DESC
LIMIT 3;
"""

top_products = pd.read_sql(query, engine)
top_products
```

```python
plt.figure(figsize=(8, 5))
plt.bar(top_products['ProductName'], top_products['TotalOrdered'], color='skyblue')
plt.title("Top 3 Most Ordered Products")
plt.xlabel("Product Name")
plt.ylabel("Total Quantity Ordered")

plt.tight_layout()
plt.show()
```
